```python
data.isnull().sum()
```

```python
data = data.drop(['id'], axis=1)
```

```python
data.head()
```

```python
encoder = LabelEncoder()

data['proto'] = encoder.fit_transform(data['proto'])
data['service'] = encoder.fit_transform(data['service'])
data['state'] = encoder.fit_transform(data['state'])
```

```python
data.head()
```

```python
pd.crosstab(data['label'], data['attack_cat'])
```

```python
labels = data[['label', 'attack_cat']]

data = data.drop(['label', 'attack_cat'], axis=1)
```

```python
X = data.select_dtypes(include=['int64','float64'])
X.head()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.head()
```

```python
X_scaled_df['label'] = labels['label'].values
X_scaled_df['attack_cat'] = labels['attack_cat'].values
X_scaled_df.head()
```

```python
X_scaled_df.to_csv('/content/drive/MyDrive/Network-Anomaly-Detection-Project/Dataset/scaled_data.csv', index=False)
```
