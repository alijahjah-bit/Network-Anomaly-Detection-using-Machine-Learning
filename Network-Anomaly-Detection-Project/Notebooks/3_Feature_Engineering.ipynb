```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

```

```python
data = pd.read_csv('/content/drive/MyDrive/Network-Anomaly-Detection-Project/Dataset/scaled_data.csv')
data.head(10)
```

```python
data['total_packets'] = data['spkts'] + data['dpkts']

data['total_bytes'] = data['sbytes'] + data['dbytes']

data['byte_ratio'] = data['sbytes'] / (data['dbytes'] + 1)

data['packets_per_sec'] = data['total_packets'] / (data['dur'] + 1)

data['bytes_per_sec'] = data['total_bytes'] / (data['dur'] + 1)

data['byte_diff'] = data['sbytes'] - data['dbytes']

data['srv_diff'] = data['ct_srv_src'] - data['ct_srv_dst']
```

```python
data.head(10)
```

```python
encoder = LabelEncoder()

for col in data.select_dtypes(include='object').columns:
    data[col] = encoder.fit_transform(data[col])
```

```python
X = data.select_dtypes(include=['int64', 'float64'])

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
X_scaled_df.to_csv("feature_data.csv", index=False)
```
